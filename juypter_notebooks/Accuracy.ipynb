```python
import pprint
import numpy as np
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
from real_time_deanonymization import get_frames
sys.path.append('../test')
import mac_randomiser_2
import accuracy_checker_2
```

```python
def run(probe_requests,database,comparison_db,delete=False):
    results, sensitivity, specificity, accuracy,metrics = [],[],[],[],[]
    for x in probe_requests:
        mac_randomiser_2.run(x,database=database,old=comparison_db)
        results = accuracy_checker_2.run(x,delete=True,database=database)
        sensitivity.append(results.sensitivity)
        specificity.append(results.specificity)
        accuracy.append(results.accuracy)
        metrics.append(results.metrics)
                    
        
    return results, sensitivity, specificity, accuracy,metrics
```

```python
%matplotlib inline
f1 = [x['f1'] for x in metrics]
p = [x['precision'] for x in metrics]
plt.plot(probe_requests, accuracy,'k--',label='Accuracy')
plt.plot(probe_requests, sensitivity, 'b',label='Sensitivity')
plt.plot(probe_requests, specificity,'y',label='Specificity')
plt.plot(probe_requests, f1, 'r',label='F1 Score')
plt.plot(probe_requests, p, 'g',label='Precision')
plt.ylim(0, 1) 
plt.title('Classifier Metrics vs Number of Probe Requests')
plt.xlabel("Number of Probe Requests")
plt.ylabel("Percentage");
# Put a legend to the right of the current axis
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
#plt.legend(bbox_to_anchor=(1, 1),
 #          bbox_transform=plt.gcf().transFigure)    
    
```
